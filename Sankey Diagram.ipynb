```python
import pandas as pd
import matplotlib as ply
import plotly.graph_objects as go

import json, urllib
import numpy as np
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
internships = pd.read_csv(r"C:\Users\User\handshake-data\Sanky test.csv")
nodes = pd.read_csv(r"C:\Users\User\handshake-data\node.csv")
paths = pd.read_csv(r"C:\Users\User\handshake-data\path.csv")
```

```python
paths = paths.replace(to_replace ="#262C46", value ="#e1e1e3")

paths.head()
```

```python
data_trace = dict(
    type='sankey',
    domain = dict(
      x =  [0,1],
      y =  [0,1]
    ),
    orientation = "h",
    valueformat = ".0f",
    node = dict(
      pad = 10,
    # thickness = 30,
      line = dict(
        color = "black",
        width = 0
      ),
      label =  nodes['Label'].dropna(axis=0, how='any'),
      color = nodes['Color']
    ),
    link = dict(
      source = paths['Source'].dropna(axis=0, how='any'),
      target = paths['Target'].dropna(axis=0, how='any'),
      value = paths['Value'].dropna(axis=0, how='any'),
      color = paths['Color'].dropna(axis=0, how='any'),
  )
)

layout = dict(
        title = "Draw Sankey Diagram from dataframes",
    height = 772,
    font = dict(
      size = 10),)

fig = dict(data=[data_trace], layout=layout)
iplot(fig, validate=False)
```
